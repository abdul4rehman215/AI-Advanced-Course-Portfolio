### Using List Comprehension

List comprehensions provide a concise way to create lists. You can use them to apply a method (like `.upper()` or `.lower()`) to each item in an existing list and create a new list with the results.

```python
# Original list
my_list = ['apple', 'Banana', 'ORANGE', 'grape']

# Convert all items to uppercase using a list comprehension
uppercase_list = [item.upper() for item in my_list]
print(f"Uppercase list: {uppercase_list}")

# Convert all items to lowercase using a list comprehension
lowercase_list = [item.lower() for item in my_list]
print(f"Lowercase list: {lowercase_list}")
```

```python
# Convert List to upper or lower case Using List Comprehension

# Original list
my_list = ['paani_wala', 'dataPeelo', 'Paanimetrics', 'PILO']

# Convert all items to uppercase using a list comprehension
uppercase_list = [item.upper() for item in my_list]

# Convert all items to lowercase using a list comprehension
lowercase_list = [item.lower() for item in my_list]

# Printing the result of uppercase list
print(f"Uppercase list: {uppercase_list}")

# Printing the result of lowercase list
print(f"Lowercase list: {lowercase_list}")
```

```python
# Convert list item to upper or lower case using the map() function with lambda

# Original list
my_list = ['paani_wala', 'dataPeelo', 'Paanimetrics', 'PILO']

# Convert all items to uppercase using the map() function with lambda
uppercase_list = list(map(lambda item: item.upper(), my_list))

# Convert all items to lowercase using the map() function with lambda
lowercase_list = list(map(lambda item: item.upper(), my_list))

# Printing the result of uppercase list
print(f"Uppercase list: {uppercase_list}")

# Printing the result of lowercase list
print(f"Lowercase list: {lowercase_list}")
```

### Using the `map()` function with `lambda`

The `map()` function applies a given function to each item of an iterable (like a list) and returns a map object (which can be converted to a list). A `lambda` function is a small anonymous function.

```python
# Original list
my_list = ['apple', 'Banana', 'ORANGE', 'grape']

# Convert all items to uppercase using map() and lambda
uppercase_list_map = list(map(lambda item: item.upper(), my_list))
print(f"Uppercase list (map): {uppercase_list_map}")

# Convert all items to lowercase using map() and lambda
lowercase_list_map = list(map(lambda item: item.lower(), my_list))
print(f"Lowercase list (map): {lowercase_list_map}")
```

### Explanation:

*   **List Comprehension (`[expression for item in iterable]`)**: This is generally preferred in Python for its readability and speed when creating new lists based on existing ones. It reads almost like natural language.
*   **`map()` with `lambda` (`list(map(function, iterable))`)**: This approach is also very efficient, especially for very large lists, and can sometimes be more concise when the transformation is a simple one-liner. The `lambda` defines a tiny function on the fly that takes an `item` and returns `item.upper()` (or `item.lower()`). The `list()` constructor converts the `map` object into a list.
