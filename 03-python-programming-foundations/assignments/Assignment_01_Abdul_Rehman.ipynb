### Question 1: Working with Lists
Lists are ordered and changeable collections. In this task, we create a list of fruits and iterate through them.

```python
# Creating a list of fruits
fruits = ["Apple", "Banana", "Cherry"]

print("Items in the fruits list:")
# Iterating through the list using a for loop
for fruit in fruits:
    print(f"- {fruit}")
```

### Question 2: Working with Tuples
Tuples are similar to lists but are immutable (cannot be changed after creation).

```python
# Creating a tuple with five random numbers
numbers = (12, 45, 7, 89, 23)

print("Items in the numbers tuple:")
for num in numbers:
    print(f"Number: {num}")
```

### Question 3: Working with Sets
Sets are unordered collections of unique items.

```python
# Creating a set of colors
colors = {"Red", "Green", "Blue"}

print("Items in the colors set:")
for color in colors:
    print(f"Color: {color}")
```

### Question 4: Working with Dictionaries
Dictionaries store data in Key-Value pairs. We will explore how to access both keys and values.

```python
# Creating a student dictionary
student = {
    "name": "Abdul Rehman",
    "age": 21,
    "grade": "A"
}

print("Displaying Dictionary Keys:")
for key in student.keys():
    print(f"Key: {key}")

print("\nDisplaying Dictionary Values:")
for value in student.values():
    print(f"Value: {value}")
```

### Question 6: List Length Check
Using the `len()` function combined with an `if` statement.

```python
# Checking the length of the fruits list
if len(fruits) > 3:
    print("You have many fruits!")
else:
    print(f"You only have {len(fruits)} fruits.")
```

### Question 7: Grading System
This program converts numerical scores into letter grades using a reusable function.

```python
def get_letter_grade(score):
    """Returns the letter grade based on a numerical score."""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Demonstration with sample values
test_scores = [95, 82, 74, 61, 45]
print("--- Sample Grade Tests ---")
for s in test_scores:
    print(f"Score: {s} -> Grade: {get_letter_grade(s)}")

# Interactive Version
print("\n--- Interactive Grade Checker ---")
try:
    user_score = float(input("Enter student numerical grade: "))
    print(f"Result: {get_letter_grade(user_score)}")
except ValueError:
    print("Invalid input! Please enter a number.")
```

### Question 8: Temperature Classifier
Classification of temperature based on ranges.

```python
def classify_temperature(temp):
    """Classifies the weather based on Celsius temperature."""
    if temp < -10:
        return "Freezing"
    elif -10 <= temp < 0:
        return "Cold"
    elif 0 <= temp < 20:
        return "Moderate"
    elif 20 <= temp < 30:
        return "Warm"
    else:
        return "Hot"

# Demonstration with sample values
temps = [-15, -5, 10, 25, 35]
print("--- Sample Temperature Tests ---")
for t in temps:
    print(f"{t}°C is classified as: {classify_temperature(t)}")

# Interactive Version
print("\n--- Interactive Weather Checker ---")
try:
    user_temp = float(input("Enter temperature in Celsius: "))
    print(f"The weather is: {classify_temperature(user_temp)}")
except ValueError:
    print("Error: Please enter a valid numerical temperature.")
```
