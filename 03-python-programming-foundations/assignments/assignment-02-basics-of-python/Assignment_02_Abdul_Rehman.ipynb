### Question 1: Vowel and Consonant Counter
We will count vowels and consonants in a string, ignoring spaces and punctuation.

```python
def count_chars(text):
    vowels = "aeiou"
    v_count = 0
    c_count = 0

    for char in text.lower():
        if char.isalpha():
            if char in vowels:
                v_count += 1
            else:
                c_count += 1
    return v_count, c_count

sample = "Hello, World!"
v, c = count_chars(sample)
print(f"Input: {sample}")
print(f"Vowels: {v}, Consonants: {c}")
```

### Question 5: BMI Calculator
Calculating Body Mass Index and classifying the result.

```python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    if bmi < 18.5:
        category = "Underweight"
    elif 18.5 <= bmi < 25:
        category = "Normal weight"
    elif 25 <= bmi < 30:
        category = "Overweight"
    else:
        category = "Obese"
    return bmi, category

# Demo
w, h = 70, 1.75
bmi_val, cat = calculate_bmi(w, h)
print(f"Weight: {w}kg, Height: {h}m")
print(f"BMI: {bmi_val:.2f}, Category: {cat}")
```

### Question 6: Character Frequency

### Question 7: Basic Email Validator
Validating an email based on the presence of '@' and specific domains.

```python
def is_valid_email(email):
    if "@" in email and (email.endswith(".com") or email.endswith(".net")):
        return True
    return False

test_email = "student@example.com"
print(f"Email: {test_email}, Valid: {is_valid_email(test_email)}")
```

### Question 9: Find Prime Numbers
Identifying the first three prime numbers from a list.

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True

numbers = [4, 7, 8, 11, 12, 13, 17, 20]
primes = [num for num in numbers if is_prime(num)]
print(f"First three primes: {primes[:3]}")
```

### Question 10: Sort Dictionaries
Sorting a list of dictionaries based on a specific key in descending order.

```python
students = [
    {"name": "John", "age": 20, "grade": 85},
    {"name": "Alice", "age": 22, "grade": 90},
    {"name": "Bob", "age": 21, "grade": 80}
]
sorted_students = sorted(students, key=lambda x: x['grade'], reverse=True)
print("Sorted by Grade (Descending):")
for s in sorted_students: print(s)
```

### Question 11: Unique Value Counter
Extracting unique items from a comma-separated user input.

```python
user_input = "10, 20, 10, 30, 40, 20"
num_list = [n.strip() for n in user_input.split(",")]
unique_vals = list(set(num_list))
print(f"Unique values: {unique_vals}")
print(f"Total unique: {len(unique_vals)}")
```

### Question 13: Merge Dictionaries with Logic
Merging two dictionaries and summing values for overlapping keys.

```python
dict1 = {'a': 5, 'b': 10, 'c': 15}
dict2 = {'b': 20, 'c': 25, 'd': 30}
merged = dict1.copy()
for k, v in dict2.items():
    merged[k] = merged.get(k, 0) + v
print(f"Merged Dictionary: {merged}")
```

### Question 16: Manual List Reversal
Reversing a list using a loop instead of built-in methods.

```python
items = [10, 20, 30, 40]
reversed_items = []
for i in range(len(items)-1, -1, -1):
    reversed_items.append(items[i])
print(f"Original: {items}")
print(f"Reversed: {reversed_items}")
```

### Question 17: Word Order Reversal
Reversing the order of words in a sentence.

```python
sentence = "Hello World"
words = sentence.split()
result = " ".join(words[::-1])
print(f"Input: {sentence} -> Output: {result}")
```

### Question 18: Palindrome Checker
Checking if a string reads the same forwards and backwards.

```python
def is_palindrome(s):
    return s == s[::-1]

word = "radar"
print(f"Is '{word}' a palindrome? {is_palindrome(word)}")
```
